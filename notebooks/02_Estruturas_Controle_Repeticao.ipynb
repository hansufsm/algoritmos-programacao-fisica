## Exercício 4.1: Classificação de Estados de Matéria

**Enunciado:**

Classifique a água em diferentes estados baseado na temperatura:
- Temperatura < 0°C: Sólido (gelo)
- 0°C ≤ Temperatura < 100°C: Líquido
- Temperatura ≥ 100°C: Gasoso (vapor)

```python
def classificar_estado_agua(temperatura):
    """
    Classifica o estado físico da água baseado na temperatura.
    
    Args:
        temperatura: temperatura em Celsius
    
    Returns:
        tuple: (estado, descricao)
    """
    if temperatura < 0:
        return ("Sólido", "Gelo")
    elif 0 <= temperatura < 100:
        return ("Líquido", "Água")
    else:
        return ("Gasoso", "Vapor")

# Testes
temperaturas_teste = [-20, 0, 25, 50, 100, 120]

print("=== CLASSIFICAÇÃO DE ESTADOS DE MATÉRIA ===")
print("Temperatura (°C) | Estado | Descrição")
print("-" * 50)

for temp in temperaturas_teste:
    estado, descricao = classificar_estado_agua(temp)
    print(f"{temp:15} | {estado:6} | {descricao}")
```

### Visualização: Aquecimento da Água e Mudanças de Estado

```python
# Simular curva de aquecimento com mudanças de estado
# Vamos simular o aquecimento de gelo até vapor

# Fases do aquecimento:
# 1. Aquecimento do gelo: -20°C até 0°C
# 2. Fusão (calor latente): permanece em 0°C
# 3. Aquecimento da água: 0°C até 100°C
# 4. Vaporização (calor latente): permanece em 100°C
# 5. Aquecimento do vapor: 100°C até 150°C

# Dados sintéticos
tempo_total = np.linspace(0, 1000, 1000)  # segundos

temperatura = np.zeros_like(tempo_total)

for i, t in enumerate(tempo_total):
    if t < 200:  # Aquecimento do gelo (-20°C a 0°C)
        temperatura[i] = -20 + (0 - (-20)) * (t / 200)
    elif t < 400:  # Fusão (plateau em 0°C)
        temperatura[i] = 0
    elif t < 800:  # Aquecimento da água (0°C a 100°C)
        temperatura[i] = 0 + (100 - 0) * ((t - 400) / 400)
    elif t < 900:  # Vaporização (plateau em 100°C)
        temperatura[i] = 100
    else:  # Aquecimento do vapor
        temperatura[i] = 100 + (150 - 100) * ((t - 900) / 100)

# Identificar estados
estado_visual = np.zeros_like(tempo_total)
cores_mapa = []

for i, temp in enumerate(temperatura):
    if temp < 0:
        estado_visual[i] = 0  # Sólido
        cores_mapa.append('lightblue')
    elif 0 <= temp < 100:
        estado_visual[i] = 1  # Líquido
        cores_mapa.append('blue')
    else:
        estado_visual[i] = 2  # Gasoso
        cores_mapa.append('red')

# Criar figura com dois painéis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), gridspec_kw={'height_ratios': [2, 1]})

# Painel 1: Curva de aquecimento
ax1.plot(tempo_total, temperatura, 'k-', linewidth=2.5)
ax1.axhline(0, color='lightblue', linestyle='--', linewidth=1.5, alpha=0.7, label='Ponto de fusão (0°C)')
ax1.axhline(100, color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='Ponto de ebulição (100°C)')

# Preencher regiões por estado
ax1.axhspan(-30, 0, alpha=0.1, color='lightblue', label='Sólido (Gelo)')
ax1.axhspan(0, 100, alpha=0.1, color='blue', label='Líquido (Água)')
ax1.axhspan(100, 160, alpha=0.1, color='red', label='Gasoso (Vapor)')

ax1.set_ylabel('Temperatura (°C)', fontsize=12, fontweight='bold')
ax1.set_title('Curva de Aquecimento da Água com Mudanças de Estado', fontsize=13, fontweight='bold')
ax1.legend(fontsize=10, loc='upper left')
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 1000)
ax1.set_ylim(-30, 160)

# Painel 2: Indicador de estado por tempo
for i in range(len(tempo_total) - 1):
    ax2.plot(tempo_total[i:i+2], [estado_visual[i], estado_visual[i+1]], 
             color=cores_mapa[i], linewidth=3)

# Rótulos dos estados
ax2.set_yticks([0, 1, 2])
ax2.set_yticklabels(['Sólido', 'Líquido', 'Gasoso'])
ax2.set_xlabel('Tempo (s)', fontsize=12, fontweight='bold')
ax2.set_title('Estado Físico ao Longo do Tempo', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='x')
ax2.set_xlim(0, 1000)
ax2.set_ylim(-0.5, 2.5)

plt.tight_layout()
plt.show()

# Tabela de resumo
print("=== MUDANÇAS DE ESTADO DA ÁGUA ===")
print()
print("Fase | Intervalo de Temperatura | Estado Físico | Processo")
print("-" * 70)
print(f"1    | -20°C a 0°C             | Sólido        | Aquecimento do gelo")
print(f"2    | 0°C (constante)         | Sólido→Líquido| Fusão (mudança de estado)")
print(f"3    | 0°C a 100°C             | Líquido       | Aquecimento da água")
print(f"4    | 100°C (constante)       | Líquido→Gasoso| Vaporização (mudança de estado)")
print(f"5    | 100°C a 150°C           | Gasoso        | Aquecimento do vapor")
```

## Exercício 5.1: Tabela de Queda Livre

**Enunciado:**

Equações:
- Velocidade: $v(t) = gt$
- Altura: $h(t) = \frac{1}{2}gt^2$

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

g = 9.81  # m/s²
tempo_maximo = 5  # segundos

# Gerar dados
tempos = list(range(0, tempo_maximo + 1))
velocidades = [g * t for t in tempos]
alturas = [0.5 * g * t**2 for t in tempos]

# Criar DataFrame
df = pd.DataFrame({
    'Tempo (s)': tempos,
    'Velocidade (m/s)': velocidades,
    'Altura (m)': alturas
})

print("=== TABELA DE QUEDA LIVRE ===")
print(df.to_string(index=False))
print()
print(f"Altura total percorrida em {tempo_maximo}s: {alturas[-1]:.2f} m")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de velocidade
ax1.plot(tempos, velocidades, 'b-o', linewidth=2, markersize=6)
ax1.set_xlabel('Tempo (s)', fontsize=11)
ax1.set_ylabel('Velocidade (m/s)', fontsize=11)
ax1.set_title('Velocidade vs Tempo - Queda Livre', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)

# Gráfico de altura
ax2.plot(tempos, alturas, 'r-s', linewidth=2, markersize=6)
ax2.set_xlabel('Tempo (s)', fontsize=11)
ax2.set_ylabel('Altura (m)', fontsize=11)
ax2.set_title('Altura vs Tempo - Queda Livre', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Exercício 5.2: Série de Leibniz para Aproximação de π

**Enunciado:**

Use a série de Leibniz para aproximar o valor de π:

$$\frac{\pi}{4} = \sum_{n=0}^{\infty} \frac{(-1)^n}{2n+1} = 1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + ...$$

Implemente um loop que pare quando a mudança for menor que $10^{-5}$.

```python
def aproximar_pi_leibniz(tolerancia=1e-5):
    """
    Aproxima π usando a série de Leibniz.
    
    Args:
        tolerancia: limite de precisão
    
    Returns:
        tuple: (valor_aproximado, numero_termos, diferenca_real)
    """
    soma = 0
    n = 0
    termo_anterior = float('inf')
    
    while True:
        termo = (-1)**n / (2*n + 1)
        soma += termo
        
        if abs(termo) < tolerancia:
            break
        
        n += 1
        
        if n > 1000000:  # Segurança
            break
    
    pi_aproximado = soma * 4
    pi_real = np.pi
    diferenca = abs(pi_aproximado - pi_real)
    
    return pi_aproximado, n + 1, diferenca

# Calcular
pi_aprox, num_termos, diferenca = aproximar_pi_leibniz()

print("=== APROXIMAÇÃO DE π - SÉRIE DE LEIBNIZ ===")
print(f"Valor aproximado de π: {pi_aprox:.10f}")
print(f"Valor real de π: {np.pi:.10f}")
print(f"Diferença: {diferenca:.10f}")
print(f"Número de termos usados: {num_termos}")
print(f"Erro percentual: {(diferenca/np.pi)*100:.6f}%")
```

### Convergência da Série

```python
# Visualizar convergência
soma = 0
n_valores = []
pi_valores = []
erros = []

for n in range(0, 10000, 100):
    soma = sum((-1)**i / (2*i + 1) for i in range(n))
    pi_aprox = soma * 4
    erro = abs(pi_aprox - np.pi)
    
    n_valores.append(n)
    pi_valores.append(pi_aprox)
    erros.append(erro)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Convergência
ax1.plot(n_valores, pi_valores, 'b-', linewidth=2, label='Aproximação')
ax1.axhline(np.pi, color='r', linestyle='--', linewidth=2, label='Valor real (π)')
ax1.set_xlabel('Número de Termos', fontsize=11)
ax1.set_ylabel('Valor aproximado', fontsize=11)
ax1.set_title('Convergência da Série de Leibniz', fontsize=12, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Erro
ax2.semilogy(n_valores, erros, 'g-', linewidth=2)
ax2.set_xlabel('Número de Termos', fontsize=11)
ax2.set_ylabel('Erro Absoluto (escala log)', fontsize=11)
ax2.set_title('Erro vs Número de Termos', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, which='both')

plt.tight_layout()
plt.show()
```

## Exercício 6.1: Produto Escalar entre Vetores

**Enunciado:**

Calcule o produto escalar entre dois vetores:

$$\vec{A} \cdot \vec{B} = \sum_{i=1}^{3} A_i B_i = A_x B_x + A_y B_y + A_z B_z$$

Teste com vetores tridimensionais.

```python
def produto_escalar(vetor_a, vetor_b):
    """
    Calcula o produto escalar entre dois vetores.
    
    Args:
        vetor_a: lista ou array com componentes de A
        vetor_b: lista ou array com componentes de B
    
    Returns:
        float: valor do produto escalar
    """
    if len(vetor_a) != len(vetor_b):
        raise ValueError("Vetores devem ter mesma dimensão")
    
    return sum(a * b for a, b in zip(vetor_a, vetor_b))

# Casos de teste
casos = [
    ([1, 0, 0], [1, 0, 0], "Vetores unitários paralelos"),
    ([1, 0, 0], [0, 1, 0], "Vetores perpendiculares"),
    ([1, 1, 0], [1, 1, 0], "Mesmo vetor"),
    ([2, 3, 4], [5, 6, 7], "Vetores arbitrários"),
]

print("=== PRODUTO ESCALAR ===")
print()

for vetor_a, vetor_b, descricao in casos:
    resultado = produto_escalar(vetor_a, vetor_b)
    print(f"{descricao}:")
    print(f"  A = {vetor_a}")
    print(f"  B = {vetor_b}")
    print(f"  A·B = {resultado}")
    print()
```

**Enunciado:**

Trabalhe com uma matriz 3×3 representando coordenadas (x, y, z) de 3 partículas.
Calcule a distância de cada uma até a origem.

```python
# Matriz de coordenadas (3 partículas, 3 coordenadas cada)
coordenadas = np.array([
    [1.0, 2.0, 2.0],      # Partícula 1: (1, 2, 2)
    [3.0, 4.0, 0.0],      # Partícula 2: (3, 4, 0)
    [1.0, 1.0, 1.0],      # Partícula 3: (1, 1, 1)
])

# Calcular distâncias até origem
distancias = np.sqrt(np.sum(coordenadas**2, axis=1))

# Criar tabela
df = pd.DataFrame({
    'Partícula': [1, 2, 3],
    'x (m)': coordenadas[:, 0],
    'y (m)': coordenadas[:, 1],
    'z (m)': coordenadas[:, 2],
    'Distância (m)': distancias
})

print("=== MATRIZ DE COORDENADAS E DISTÂNCIAS ===")
print(df.to_string(index=False))
print()
print(f"Partícula mais próxima da origem: {np.argmin(distancias) + 1} (d = {np.min(distancias):.3f} m)")
print(f"Partícula mais distante da origem: {np.argmax(distancias) + 1} (d = {np.max(distancias):.3f} m)")
```

## Exercício Proposto 3: Validação Repetida de Entrada

**Desafio:**

Implemente um programa que:
1. Solicita um valor ao usuário
2. Valida se é positivo
3. Repete até obter entrada válida
4. Calcula a raiz quadrada do valor

Use um loop `while` com validação.

```python
def soma_vetores(vetor_a, vetor_b):
    """
    Soma dois vetores componente a componente.
    
    Args:
        vetor_a: lista ou array com componentes de A
        vetor_b: lista ou array com componentes de B
    
    Returns:
        ndarray: vetor soma C = A + B
    """
    return np.array(vetor_a) + np.array(vetor_b)

def magnitude_vetor(vetor):
    """
    Calcula a magnitude (módulo) de um vetor.
    
    Args:
        vetor: lista ou array com componentes
    
    Returns:
        float: magnitude do vetor
    """
    return np.sqrt(np.sum(np.array(vetor)**2))

def angulo_entre_vetores(vetor_a, vetor_b):
    """
    Calcula o ângulo (em graus) entre dois vetores.
    
    Args:
        vetor_a, vetor_b: listas ou arrays
    
    Returns:
        float: ângulo em graus
    """
    va = np.array(vetor_a)
    vb = np.array(vetor_b)
    cos_angulo = np.dot(va, vb) / (magnitude_vetor(va) * magnitude_vetor(vb))
    # Clampar para evitar erros numéricos
    cos_angulo = np.clip(cos_angulo, -1, 1)
    return np.degrees(np.arccos(cos_angulo))

# Casos de teste
casos = [
    ([3, 0], [0, 4], "Vetores perpendiculares"),
    ([1, 1], [2, 1], "Vetores arbitrários"),
    ([2, 0], [1, 0], "Vetores colineares"),
]

print("=== SOMA E MAGNITUDE DE VETORES ===")
print()

for vetor_a, vetor_b, descricao in casos:
    vetor_c = soma_vetores(vetor_a, vetor_b)
    mag_a = magnitude_vetor(vetor_a)
    mag_b = magnitude_vetor(vetor_b)
    mag_c = magnitude_vetor(vetor_c)
    angulo = angulo_entre_vetores(vetor_a, vetor_b)
    
    print(f"{descricao}:")
    print(f"  A = {vetor_a} (|A| = {mag_a:.3f})")
    print(f"  B = {vetor_b} (|B| = {mag_b:.3f})")
    print(f"  C = A + B = {list(vetor_c)} (|C| = {mag_c:.3f})")
    print(f"  Ângulo entre A e B: {angulo:.2f}°")
    print()

# Visualização: Regra do Paralelogramo
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for idx, (vetor_a, vetor_b, titulo) in enumerate(casos):
    ax = axes[idx]
    
    vetor_c = soma_vetores(vetor_a, vetor_b)
    
    # Plotar vetores como setas
    # Vetor A (origem)
    ax.arrow(0, 0, vetor_a[0], vetor_a[1], head_width=0.2, head_length=0.15, 
             fc='red', ec='red', linewidth=2.5, label='A', zorder=3)
    
    # Vetor B (origem)
    ax.arrow(0, 0, vetor_b[0], vetor_b[1], head_width=0.2, head_length=0.15, 
             fc='blue', ec='blue', linewidth=2.5, label='B', zorder=3)
    
    # Vetor soma C (do ponto de vista da regra do paralelogramo)
    ax.arrow(0, 0, vetor_c[0], vetor_c[1], head_width=0.2, head_length=0.15, 
             fc='green', ec='green', linewidth=3, label='C = A + B', zorder=4)
    
    # Linhas do paralelogramo (tracejadas)
    ax.plot([vetor_a[0], vetor_c[0]], [vetor_a[1], vetor_c[1]], 'b--', linewidth=1, alpha=0.6, zorder=1)
    ax.plot([vetor_b[0], vetor_c[0]], [vetor_b[1], vetor_c[1]], 'r--', linewidth=1, alpha=0.6, zorder=1)
    
    # Pontos notáveis
    ax.scatter([0], [0], c='black', s=100, marker='o', zorder=5, label='Origem')
    ax.scatter([vetor_c[0]], [vetor_c[1]], c='green', s=100, marker='*', zorder=5)
    
    # Formatação
    ax.set_xlim(-1, max(vetor_a[0], vetor_b[0], vetor_c[0]) + 1)
    ax.set_ylim(-1, max(vetor_a[1], vetor_b[1], vetor_c[1]) + 1)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(0, color='k', linewidth=0.5)
    ax.axvline(0, color='k', linewidth=0.5)
    ax.set_xlabel('X', fontsize=11)
    ax.set_ylabel('Y', fontsize=11)
    ax.set_title(titulo, fontsize=12, fontweight='bold')
    ax.legend(fontsize=10, loc='best')

plt.suptitle('Soma de Vetores - Regra do Paralelogramo', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
